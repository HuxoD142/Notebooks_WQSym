Il faut ajouter au fichier les class et type pour pouvoir afficher les ordres correctement. Puis après même pour le strongly fatter c'est encore autre chose.


```python
from packed_words import *
```

```python
def pw_class(pw):
    n=len(pw)
    m=max(pw)
    cl=[0]*m
    for i,x in enumerate(pw):
        if cl[x-1]==0:
            cl[x-1]=max(cl)+1
    return [cl[i-1] for i in pw]
```

```python
def pw_by_class(lw):
    d=dict()
    for pw in lw:
        cl=str(pw_class(pw))
        if cl in d:
            d[cl]+=[pw]
        else:
            d[cl]=[pw]
    return d
```

```python
def LWO_multi_plot(n):
    pwn=pw_by_class(PackedWords(n))
    res=[]
    for c in pwn:
        elms = range(len(pwn[c]))
        fcn = lambda p,q : pwn[c][q] in pwn[c][p].left_weak_order_smaller()
        d={i:str(pwn[c][i]) for i in elms}

        res.append(Poset((elms,fcn)).plot(element_labels=d, vertex_size=20))

    return res
```
